```python
# Dependencies
from api_keys import api_key
from pprint import pprint
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import requests
import time
import json
import os
import re
import pylab

url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"
```

```python
# Store a search term
query = "rock music"
rock_yearly_count={}
time_sec=0
for year in range(1921,2021):
    try:
        time_sec+=1
        # Search for articles published between a begin and end date
        begin_date=f'{year}0101'
        end_date=f'{year}1231'
        query_url = f"{url}api-key={api_key}&q={query}&facet_fields=source&facet=true&begin_date={begin_date}&end_date={end_date}"
        temp = requests.get(query_url).json()
        print(f'processing year {year}')
        rock_yearly_count[year]=temp['response']['facets']['source']['terms'][0]['count']
        if time_sec>9:
            time.sleep(60)
            time_sec=0
    except:
        print(f'year {year} error')
        time.sleep(60)
        continue
rock_yearly_count
```

```python
# Store a search term
query = "pop music"
pop_yearly_count={}
time_sec=0
for year in range(1921,2021):
    try:
        time_sec+=1
        # Search for articles published between a begin and end date
        begin_date=f'{year}0101'
        end_date=f'{year}1231'
        query_url = f"{url}api-key={api_key}&q={query}&facet_fields=source&facet=true&begin_date={begin_date}&end_date={end_date}"
        temp = requests.get(query_url).json()
        print(f'processing year {year}')
        pop_yearly_count[year]=temp['response']['facets']['source']['terms'][0]['count']
        if time_sec>9:
            time.sleep(60)
            time_sec=0
    except:
        print(f'year {year} error')
        time.sleep(60)
        continue
pop_yearly_count
```

```python
# Store a search term
query = "hip hop"
hiphop_yearly_count={}
time_sec=0
for year in range(1921,2021):
    try:
        time_sec+=1
        # Search for articles published between a begin and end date
        begin_date=f'{year}0101'
        end_date=f'{year}1231'
        query_url = f"{url}api-key={api_key}&q={query}&facet_fields=source&facet=true&begin_date={begin_date}&end_date={end_date}"
        temp = requests.get(query_url).json()
        print(f'processing year {year}')
        hiphop_yearly_count[year]=temp['response']['facets']['source']['terms'][0]['count']
        if time_sec>9:
            time.sleep(60)
            time_sec=0
    except:
        print(f'year {year} error')
        continue
hiphop_yearly_count
```

```python
%matplotlib notebook
rock_lists = sorted(rock_yearly_count.items())
pop_lists = sorted(pop_yearly_count.items())
hiphop_lists = sorted(hiphop_yearly_count.items())

x_rock,y_rock=zip(*rock_lists)
x_pop,y_pop=zip(*pop_lists)
x_hiphop,y_hiphop=zip(*hiphop_lists)

plt.plot(x_rock, y_rock)
plt.plot(x_pop, y_pop)
plt.plot(x_hiphop, y_hiphop)
plt.ylabel('Article Counts')
plt.xlabel('Year')
plt.legend(['Rock','Pop','Hip Hop'],
          title="Genres",
          loc="upper left")
plt.title("Article Counts on Music Genres")
plt.show()
```

```python
rock_yearly_df=pd.DataFrame(rock_yearly_count.items(),columns=['Year', 'Rock'])
pop_yearly_df=pd.DataFrame(pop_yearly_count.items(),columns=['Year', 'Pop'])
hiphop_yearly_df=pd.DataFrame(hiphop_yearly_count.items(),columns=['Year', 'Hip Hop'])
nyt_genre_df=pd.merge(rock_yearly_df,pop_yearly_df)
nyt_genre_df=pd.merge(nyt_genre_df, hiphop_yearly_df, how='outer')
nyt_genre_df.to_csv("Output_data/nyt_genre.csv")
nyt_genre_df
```
